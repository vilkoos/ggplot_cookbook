# 2 - Simplified forms

It is possible to produce ggplot's using simplified syntax (i.e. it is possible to produce the same plots with less typing).  
To start with, arguments not need to be named when they are put at the default position.
- the data=dataframe --- can be --- ggplot(dataframe, ...)
- the mapping=aes() --- can be -- aes() at the second position
- the first argument within the aes() is considered to be the x=... the second the y=...

Here is an example

```python
figure = ggplot(dia, aes('carat', 'price')) # default rule: first the dataframe, then within the aes the x and y
figure += geom_point()                      # f = f + g can be shortend to f += g
print(figure)
```

It is possible to merge the first two lines

```python
figure = ggplot(dia, aes('carat', 'price')) + geom_point()
print(figure)
```

There is no iron rule that plots need to be stored in a variable like 'figure'  
The below works just as well

```python
ggplot(dia, aes('carat', 'price')) + geom_point()
```

Not everything always fits on one line,  
In that case you can use this form. (**NOTE no spaces after the \** )

```python
ggplot(dia, aes('carat', 'price')) + \
   geom_point() + \
   geom_point(size=3, color='orange', alpha=0.3) 
```

When you want to avoid the \ at the end of the lines, use this form

```python
(
ggplot(dia, aes('carat', 'price')) + 
   geom_point() + 
   geom_point(size=3, color='orange', alpha=0.3)
)
```

# 3 - form used here

This cookbook:
- will always name arguments (according to the [Zen of Python](https://www.python.org/dev/peps/pep-0020/) explicit is better than implicit)
- will give each layer (base, geom etc) its own line
- will not store the result in a variable like figure unless necessary.  
- will use + \ at the end of a line if necessary
