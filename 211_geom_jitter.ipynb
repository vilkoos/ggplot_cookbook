# ggplot 211 - How to make a jitter plot

This notebook shows how to make a jitter plot with geom_jitter to prevent overplotting.   
It also explains why you would want to use jitter plots.

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
dia = pd.read_csv('./dat/dia.csv')
```

# 2 - example : dot plot price by carat

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point()
```

This figure shows 500 solid black points. Points that are located in the same position overlap.  
#### problem: It is impossible to see if there are 1, 2 or more points underneath a point (overplotting problem)
Setting the alpha to 0.2 and making the points smaller is one way to solve this "hidden" points problem.  
(adding color for different classes of points might also help)

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point(alpha=0.5, size=0.5)
```

Adding a some random jitter to each x and y coordinate often has a positive effect (i.e. prevents overplotting).

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_jitter(alpha=0.5, size=0.5, width=0.1, height=100)
# width=0.1 adds a random jitter in the range -0.1 to 0.1 to each x coordinate 
# height=100 adds random jitter in the range -100 to 100 to each y coordinate
# because of this jitter clusters of overlapping points are broken up.
```

# zoom in

xlim and ylim can be used to zoom in on a certain region of the plot.  
If we zoom in on the points-cloud in the lower left corner, the cloud dissolves into points.

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_jitter(alpha=0.5, size=0.5, width=0.1, height=100) + \
    xlim([0.15,0.5]) + ylim([250,1500])
```
