# ggplot 220 - How to make a line plot

This notbook shows how to make line plots with geom_line

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
dia = pd.read_csv('./dat/dia.csv')
```

# 2 - example : line plot price by carat

```python
figure = ggplot(mapping=aes(x='carat', y='price'), data=dia)
figure = figure + geom_line()
print(figure)
```

# Aesthetics

According to the [documentation](http://ggplot.yhathq.com/docs/geom_line.html) geom_line has aesthetics

- x -------- (x position)
- y -------- (y position)
- alpha ---- (transparency)
- color ---- (color)
- linetype - (either "solid", "dashed" or "dotted" - solid is the default-)
- size ----- (how thick is the line)

```python
#-- make a simple data set with four points----------------
df = pd.DataFrame([[0,1],[1,1],[2,0],[3,2]], columns=['X_axis', 'Y_axis'])
df
```

```python
# --- show the effect of the aesthetics ---------
ggplot(mapping=aes(x='X_axis',y='Y_axis'), data=df) + \
    geom_line(linetype='dotted', size=3, color='red')
```

# combine geom_line with geom_point

Each geom creates a layer. We can put layers on top of each other. Here is an example.

```python
ggplot(mapping=aes(x='X_axis',y='Y_axis'), data=df) + \
    geom_line(linetype='dashed', size=5, color='red', alpha=0.3) + \
    geom_line(linetype='dotted', size=1, color='black') + \
    geom_point(size=10, color='blue',alpha=0.3)
```

# create one line per "color"

We can use color in the aes() to create one line for each vaule of color (e.g. one line for each way a diamond is cut). 

```python
figure = ggplot(mapping=aes(x='carat', y='price',color='cut'), data=dia)
figure = figure + geom_line()
print(figure)
```

Now we can see the effect of cut on the relationship between carat and price.
