# ggplot 241 - How to make a violin plot
This notebook shows how to make a violin plot with geom_violin.

In many cases a violin plot is a better alternative for a boxplot.

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
dia = pd.read_csv('./dat/dia.csv')
```

# 2 - example : violin plot of diamond prices by cut

```python
figure = ggplot(mapping=aes(x='cut', y='price'), data=dia)
figure = figure + geom_violin()
print(figure)
```

This shows how the prices of different cuts of diamonds are distributed.  
(the broader a figure is the more cases at that price level there are).

Lets compare this with a boxplot of the same data

```python
figure = ggplot(mapping=aes(x='cut',y='price'), data=dia)
figure = figure + geom_boxplot()
print(figure)
```

Both plots show the same overall pattern (show the same distribution).  
Howevver there is an important difference
- the violin plot shows a smooth outline of the distribution,  
- the boxplot shows a chuncky outline (i.e the median, the Q1, Q2 (median) the Q3 and the outliers.)

The violin plot gives a better insight into the distribution.  
The boxplot is better at showing outliers.

See also Wikipedia on [violin plots](https://en.wikipedia.org/wiki/Violin_plot)


## Aesthetics

According to the [documentation](http://ggplot.yhathq.com/docs/geom_density.html) geom_density has aesthetics.
- x -------- (variable to plot along the x-axis) Note should be categorical (i.e. nominal or ordinal)
- y -------- (variable to plot along the y-axis) Note should be continuous (i.e. interval or ratio)
- alpha ---- (transparency)
- color ---- (color of the border)
- fill ----- (color of the violin)
- linetype - (either "solid", "dashed" or "dotted" - solid is the default-)
- size ----- (how thick is the boder line)
- weight --- (not implemented in plotnine?)

# 3 - violin plot of diamond prices

```python
ggplot(mapping=aes(x='cut', y='price'), data=dia) + \
   geom_violin(size=2, color='red', linetype='dotted', fill='yellow', alpha=0.6)
```
