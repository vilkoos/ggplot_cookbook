# ggplot 210 - How to make a points plot

This notebook shows how to make a point plot (aka dotplot or scatterplot) with geom_point.  

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
dia = pd.read_csv('./dat/dia.csv')
```

# 2 - example : dot plot price by carat

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point()
```

# Aesthetics

According to the [documentation](http://ggplot.yhathq.com/docs/geom_point.html) geom_point has aesthetics

- x -------- (x position)
- y -------- (y position)
- alpha --- (transparency)
- color ---- (color)
- fill ------- (color of the inside of the plot)
- size ----- (how large is the diameter of the point)
- shape ---- (the shape of the point)
- stroke --- (how thick is the border of the point)

## Plotnine shapes:
- 'o' --- a solid dot (the default)
- '^' --- a triangle with the point up
- 'v' --- a triangle with the point down
- 'D' --- a diamond
- 's' --- a square
- 'p' --- a pentagram with the point up
- '\*' -- an astrix (*)
- '_' --- a horizontal dash
- '|' --- a vertical dash
- '8' --- an octagon 

NOTE: ggplot2 in R has other shapes (see [cheatsheet](https://www.rstudio.com/wp-content/uploads/2015/03/ggplot2-cheatsheet.pdf))

# 3 -  effect of aesthetics

```python
#-- make a simple data set with four points for the demo ----------------
df = pd.DataFrame([[0,1],[1,1],[2,0],[3,2]], columns=['X_axis', 'Y_axis'])
df
```

```python
# --- show the effect of the aesthetics ---------
ggplot(data=df, mapping=aes(x='X_axis',y='Y_axis')) + \
    geom_point(size=10,fill='yellow',stroke=5,color='red')
# size = the diameter of the point
# fill = color of the inside of the point
# stroke = the thickness of the border of the point
# color = color of the border of the point
```

```python
# --- show the effect of the aesthetics ---------
ggplot(data=df, mapping=aes(x='X_axis',y='Y_axis')) + \
    geom_point(size=10, color='red',shape='p') + \
    geom_point(size=15, color='yellow',shape='o', alpha=0.5) + \
    geom_line(linetype='dotted', color='blue')
```

```python
# --- show the effect of the octagon aesthetics ---------
ggplot(data=df, mapping=aes(x='X_axis',y='Y_axis')) + \
    geom_point(size=30, color='blue',shape='8')
```

Only one octagon is completely visible.   
Correct that by showing a slightly wider grid by setting xlim() and ylim()

```python
# --- set the shown area with xlim() and ylim() ---------
ggplot(data=df, mapping=aes(x='X_axis',y='Y_axis')) + \
    geom_point(size=30, color='blue',shape='8') + \
    xlim(-1,4) + ylim(-1,3)
```

# show all data in one plot

We can use color= and shape= in the aes() to make the diamond color and shape visibel in the plot. 

```python
dia = pd.read_csv('./dat/dia.csv')
ggplot(data=dia, mapping=aes(x='carat', y='price',color='color',shape='cut')) + \
    geom_point()
```
