# ggplot 232 - How to make a density plot
This notebook shows how to make a density plot with geom_density.

## density plots can only be made of continious numeric variables (i.e. interval or ratio).  

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
dia = pd.read_csv('./dat/dia.csv')
```

# 2 - example : density plot of diamond prices

```python
figure = ggplot(mapping=aes(x='price'), data=dia)
figure = figure + geom_density()
print(figure)
```

Lets compare this with a histogram of the same data

```python
figure = ggplot(mapping=aes(x='price'), data=dia)
figure = figure + geom_histogram(binwidth=500)
print(figure)
```

Both plots show the same overall pattern (show the same distribution).  
Howevver there are important differences
- the density plot shows a smooth outline of the distribution
- the Y-axis of the density plot does not show the counts but the density (hence the name density plot).

The density numbers are calculated in such a way that the total erea under the curve is exactly 1.   
Hence the density plot shows the probability distribution of a [continuous variable](https://www.khanacademy.org/math/statistics-probability/random-variables-stats-library/discrete-and-continuous-random-variables/v/probability-density-functions).

## Aesthetics

According to the [documentation](http://ggplot.yhathq.com/docs/geom_density.html) geom_density has aesthetics.
- x -------- (variable to plot along the x-axis)
- alpha ---- (transparency)
- color ---- (color of the border)
- fill ----- (color of the bar)
- linetype - (either "solid", "dashed" or "dotted" - solid is the default-)
- size ----- (how thick is the boder line)
- weight --- (not implemented in plotnine?)

# 3 - density plot of diamond prices

```python
ggplot(mapping=aes(x='price'), data=dia) + \
   geom_density(size=2, color='red', linetype='dotted', fill='yellow', alpha=0.6)
```

# 4 - price density plots per diamond cut

```python
ggplot(mapping=aes(x='price', color='cut', fill='cut'), data=dia) + \
   geom_density(alpha=0.3)
```
