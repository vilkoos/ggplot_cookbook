# ggplot 240 - How to make a box plot
This notebook shows how to make a box plot with geom_boxplot.

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
dia = pd.read_csv('./dat/dia.csv')
```

# 2 - example box plot

```python
ggplot(data=dia, mapping=aes(x='cut', y='price')) + \
    geom_boxplot()
```

For each of the values of cut a [boxplot](https://en.wikipedia.org/wiki/Box_plot) is drawn.  
The line in the middle of each box represents the median.  
The lower box edge is the first quartile, and the upper edge shows the third quartile.   
All the points above the 95 percentile are shown as points.  
All the points below the 5 percentile are shown  as points.  
These points are probably outliers.

Boxplots are usefull for detecting outliers.  
They are also usefull for checking if a distribution is symetric.  
In the symettric case the median is in the middle of the box, and the wiskers are of equal length.

**Tip** By adding a fill='cut' we get a colored diagram and a legend.

```python
ggplot(data=dia, mapping=aes(x='cut', y='price',  fill='cut')) + \
    geom_boxplot()
```

The flipped boxplot is a little easier to read.

```python
ggplot(data=dia, mapping=aes(x='cut', y='price',  fill='cut')) + \
    geom_boxplot() + \
    coord_flip()
```

For each value of the x varaible a boxplot is drawn.  
That obviously does not work for continious (i.e. interval or ratio) variables.  
Continious variables can have infinite many different values, we can not draw infinite many boxplots.

When we group the values we could draw a boxplot for each group.  
When we do not group a continious variable all values will be put into one group, so one boxplot is drawn.  
See the example below.

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_boxplot()
```

The x-axis label 'carat' is not meaningfull, we can remove that with xlab("").

## Aesthetics

According to the [documentation](https://plotnine.readthedocs.io/en/stable/generated/plotnine.geoms.geom_boxplot.html#plotnine.geoms.geom_boxplot) geom_density has aesthetics.
- x -------- (variable to plot along the x-axis) usally a categorical variablr
- y -------- (variable to plot along the y-axis) a continious variable
- alpha ---- (transparency)
- color ---- (color of the border)
- fill ----- (color of the bar)
- linetype - (either "solid", "dashed" or "dotted" - solid is the default-)
- size ----- (how thick is the boder line)

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
   geom_boxplot(size=2, color='red', linetype='dotted', fill='yellow', alpha=0.6) + \
   xlab("")
```

NOTE the outliers can have their own aestetics
- outlier_color ---- (the color of the outlier)
- outlier_shape ---- (the shape of the outlier)
- outlier_size ----- (the size of the outlier)

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
   geom_boxplot(fill='yellow', alpha=0.3, outlier_color = "red", outlier_shape = '_', outlier_size=5) + \
   xlab('-')
```
