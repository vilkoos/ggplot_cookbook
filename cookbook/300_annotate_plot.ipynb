# ggplot 300 - How to annotate a plot

This notebook shows how we can add a title with ggtitle.  
It also shows how to add custom labels for the X and Y-axis with xlab and ylab.  

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
dia = pd.read_csv('./dat/dia.csv')
```

# 2 - adding a title and custom x an y labels 

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point(size=0.5, color='blue') + \
    ggtitle('diamond price by carat') + \
    xlab('weight in carat') + ylab('price in US-dollars')
```
