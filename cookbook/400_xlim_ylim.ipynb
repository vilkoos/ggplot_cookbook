# ggplot 400 - how to limit the plotted area with xlim and ylim

This notebook shows how to xlim and ylim to limit the nplotted area.  

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
dia = pd.read_csv('./dat/dia.csv')
```

# 2 - example : dot plot price by carat

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point()
```

# 3a - limit the plotted x range

plot only points between 1 en 2 carat.

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point() + \
    xlim([1,2])
```

# 3b - limit the plotted y range

plot only points between 5000 en 10000 dollar.

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point() + \
    ylim([5000,10000])
```

# 3c - limit both the x and the y range

```python
ggplot( data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point() + \
    xlim([1,2]) + ylim([5000,10000])
```
