# ggplot 410 - customize continuous x and y axis scales

With scales_x_continuous we can set:
- the breaks along the x-axis (where the ticks are)
- the labels of the breaks
- the limits between which data is plotted

scales_y_continuous does the same for the y-axis

The axis labels can be customized with xlab() and ylab()

By setting the x and y limits we can zoom in on a sub-area of the plot.

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
dia = pd.read_csv('./dat/dia.csv')
```

# 2 - the un-customized plot

Without customization the plot looks like this

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point(size=0.3, color='blue')
```

# 2.1 - customize the x scale

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point(size=0.3, color='blue') + \
    scale_x_continuous(breaks=[0,0.5,1,1.5,2,2.5,3], 
                       limits=[-1,4],
                       labels=['0/2','1/2','2/2','3/2','4/2','5/2','6/2']) + \
    xlab("weight in carats - (1 carat=200 milligrams)")
```

# 2.2 - customize the y scale

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point(size=0.3, color='blue') + \
    scale_y_continuous(breaks=[2500,5000,7500,10000], 
                       limits=[2500,10000],
                       labels=['2.5','5','7.5','10']) + \
    ylab("price (in units of 1000 US dollar)")
```

Note that all points with an y value above 10000 are not visible.

# 2.3 - zoom in on a sub-area

By combining scale_x and scale_y we can zoom in on particular areas of the plot.

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point(size=0.3, color='blue') + \
    scale_y_continuous(limits=[2500,5000]) + \
    scale_x_continuous(limits=[0.5,1.25])
```

The same can be done using xlim() and ylim()

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point(size=0.3, color='blue') + \
    ylim(2500,5000) + \
    xlim(0.5,1.25)
```

# 2.4 - zoom out
We can use the same technique to zoom out (just set the limits to a wider range)

```python
ggplot(data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point(size=0.3, color='blue') + \
    ylim(-5000,25000) + \
    xlim(-1,5)
```
