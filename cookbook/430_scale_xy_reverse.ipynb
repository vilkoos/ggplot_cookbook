# ggplot 430 - reverse x and y axis scales

With scales_x_reverse we can let the scale go from right to left (instead of left to right)
With scales_y_reverse we can let the scale go from up to down (instead of down to up).
Also we can set
- the breaks along the x-axis (where the ticks are)
- the labels of the breaks
- the limits between which data is plotted

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
dia = pd.read_csv('./dat/dia.csv')
```

# 2 - the un-customized plot

Without customization the plot looks like this

```python
ggplot( data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point(size=0.3, color='blue')
```

# 2.1 - reverse the x scale

```python
ggplot( data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point(size=0.3, color='blue') + \
    scale_x_reverse()
```

In scale_x_reverse() we can (re)set
- the breaks
- the labels
- the x limits

```python
ggplot( data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point(size=0.3, color='blue') + \
    scale_x_reverse(breaks=[0,0.5,1,1.5,2,2.5,3], 
                       limits=[-1,4],
                       labels=['0/2','1/2','2/2','3/2','4/2','5/2','6/2']) + \
    xlab("weight in carats - (1 carat=200 milligrams)")
```

# 2.2 - reverse the y scale

This works the same as the x reverse.

```python
ggplot( data=dia, mapping=aes(x='carat', y='price')) + \
    geom_point(size=0.3, color='blue') + \
    scale_y_reverse(breaks=[2500,5000,7500,10000], 
                       limits=[2500,10000],
                       labels=['2.5','5','7.5','10']) + \
    ylab("price (in units of 1000 US dollar)")
```
