# ggplot 450 - How to customize discrete color scales

The aesthetic color=... can give different colors to data values.  
With scale_color_brewer we can control the used color scheme.

Note: 
- use scale_color_brewer for discrete (nominal or ordinal) variables
- use scale_color_gradient for  continuous (interval or ratio) variables.

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
dia = pd.read_csv('./dat/dia.csv')
```

# 2 - un-customized plot

```python
ggplot(data=dia, mapping=aes(x='carat', y='price', color='cut')) + \
    geom_point()
```

We can change the used colors using the color brewer approach.  
Have a look at the [color brewer](http://colorbrewer2.org/) documentation first.

Color brewer has designed pallets for three types of data.
- type = sequential (i.e. ordinal data with gradual differences)
- type = diverging (i.e. ordinal with dramatic contrasts between the extremes)
- type = qualitative (i.e. nominal data )

# 3 - sequential

```python
ggplot(data=dia, mapping=aes(x='carat', y='price', color='cut')) + \
    geom_point() + \
    scale_color_brewer(type='seq')
```

The order is expressed as a movement from very light blue to very dark blue.

There are a number of different sequential color pallets.  
Palette=1 is the default (shown above).  
Palette=7 is shown below.

```python
ggplot(data=dia, mapping=aes(x='carat', y='price', color='cut')) + \
    geom_point() + \
    scale_color_brewer(type='seq', palette=7)
```

# 4 -  diverging 

```python
ggplot(data=dia, mapping=aes(x='carat', y='price', color='cut')) + \
    geom_point() + \
    scale_color_brewer(type='div')
```

Note that here two colors are used, the contrast between the extremes of the scale are very clear.  
We can again use several palettes.

```python
ggplot(data=dia, mapping=aes(x='carat', y='price', color='cut')) + \
    geom_point() + \
    scale_color_brewer(type='div', palette=4)
```

# 5 - qualitative

```python
ggplot(data=dia, mapping=aes(x='carat', y='price', color='cut')) + \
    geom_point() + \
    scale_color_brewer(type='qual', palette=6)
```

Here there is no "ordering" of the colors.  
The colors are chosen to make it easy to distinguish the categories.
