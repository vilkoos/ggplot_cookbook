# ggplot 460 - How to customize continuous color scales

The aesthetic color=... can give different colors to data values.  
With scale_color_gradient we can control the used color scheme.

Note: 
- use scale_color_brewer for discrete (nominal or ordinal) variables
- use scale_color_gradient for  continuous (interval or ratio) variables.  

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
dat = pd.read_csv('./dat/income.csv')
```

```python
dat.head()
```

For 180 countries three continuous variables are given:
- INC - income per capita
- LE - life expectancy in years
- IR - internet use rate (% of population with internet access)

```python
dat.info()
```

```python
dat.describe().round(1)
```

# 2 - un-customized plot

```python
ggplot(data=dat, mapping=aes(x='INC', y='LE', color='IR')) + \
    geom_point() + \
    scale_x_log10()
```

We can change the used colors using the scale_color_gradient.  
With that we can choose the colors in our grandient scale.

# 3 - customized plots

```python
ggplot(data=dat, mapping=aes(x='INC', y='LE', color='IR')) + \
    geom_point() + \
    scale_x_log10() + \
    scale_color_gradient(low='white', high='red')
```

The order is expressed as a movement from very light blue to very dark red.

You can use every combination of colors.

```python
ggplot(data=dat, mapping=aes(x='INC', y='LE', color='IR')) + \
    geom_point() + \
    scale_x_log10() + \
    scale_color_gradient(low='blue', high='red')
```

[Hexadecimal color codes](http://htmlcolorcodes.com/color-chart/flat-design-color-chart/) can be used also.  
see also the html [color picker](https://www.w3schools.com/colors/colors_picker.asp)

```python
ggplot(data=dat, mapping=aes(x='INC', y='LE', color='IR')) + \
    geom_point() + \
    scale_x_log10() + \
    scale_color_gradient(low='#5011D1', high='#05D9F6')
```
