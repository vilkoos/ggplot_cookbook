# ggplot 500 - How flip the x and y coordinates

This notebook shows how to switch the x and y axis.  
Often flipped plot are better to read. 

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
dia = pd.read_csv('./dat/dia.csv')
```

# 2 - un-flipped boxplot

```python
ggplot(data=dia, mapping=aes(x='cut', y='price',  fill='cut')) + \
    geom_boxplot()
```

# 3 - flipped boxplot

Flipp the boxplot with coord_()

```python
ggplot(data=dia, mapping=aes(x='cut', y='price',  fill='cut')) + \
    geom_boxplot() + \
    coord_flip()
```

# 4 - flipped bar-chart

```python
ggplot(data=dia, mapping=aes(x='cut', fill='cut')) + \
    geom_bar() + \
    coord_flip()
```

# 5 -  flipped histogram

```python
ggplot(data=dia, mapping=aes(x='price')) + \
   geom_histogram(bins=20, fill='teal') + \
   coord_flip()
```

## triple flip
Remember, just because you can, does not mean you must.

```python
ggplot(data=dia, mapping=aes(x='price')) + \
   geom_histogram(bins=20, fill='teal') + \
   scale_x_reverse() + \
   scale_y_reverse() + \
   coord_flip()
```
