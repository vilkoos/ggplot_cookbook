# ggplot 700 - use themes

It is possible to specify which background and grid lines, fonts etc you want to use.  

A set of such specifications is called a theme.  
Themes do not change the geom's, they do change the "background-grid" on which the geom's are displayed.  
It is possible to define your own themes (see the theme documentation)

plotnine comes with a set of predefined themes.

# 0 - setup notebook

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 
```

# 1 - get some data

```python
car = pd.read_csv('./dat/mtcars.csv')

def make_cyl_ordinal(cyl):
    return 'c'+str(cyl)
car['cylOrd'] = car['cyl'].apply(make_cyl_ordinal)
```

# 2 - comparison 

```python
ggplot(data=car, mapping=aes(x='wt', y='mpg', color='cylOrd') ) + \
    geom_line() + \
    ggtitle("default (theme_gray)") + \
    theme_gray()
```
