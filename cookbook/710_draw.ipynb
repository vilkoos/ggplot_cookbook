# ggplot 710 - set plot size using draw() 

Plotnine has its own ggplot format for figures.  
It is possible to convert this format to standard matplotlib.  
The plotnine draw() method implements this conversion.  

The converted plot can be resized with the matplotlib set_size_inches() method.   
This lets us control the sizes of the plots in our notebook.

# 0 - setup notebook

Note that we import matplotlib.
We also set the notebook to show matplotlib plots inline.

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import * 

#---- import matplotlib --------------
import matplotlib
import matplotlib.pyplot as plt
#---- display matplotlib plot inline in this notebook -------
%matplotlib inline
```

# 1 - get some data

```python
dia = pd.read_csv('./dat/dia.csv')
```

# 2 - set size of plots 

Suppose we would like our plots in the size of inches 3 by 2 .   
To do that:
- store the plot in a variable fig.
- convert fig to mathplotlib with draw() 
- use the set_size_inches(3,2) matplotlib method on fig

```python
fig = ggplot(data=dia, mapping=aes(x='color', fill='cut')) + geom_bar()
fig.draw().set_size_inches(3,2)
```

Note that fig.set_size_inches(3,2) will not work  
(since fig is in ggplot format not in matplotlib format)

```python
fig.set_size_inches(5,4)
```

In this way we can control the size of all of our plots.

```python
fig2  = ggplot( data=dia, mapping=aes(x='carat', y='price', color='color')) + \
            geom_point(size=2, alpha=0.4 )
fig2.draw().set_size_inches(3,2)
```

# appendix - ggplot .draw() and .save methods


### draw()
- Render the complete plot and return the matplotlib figure

### save()
(filename=None, format=None, path=None, width=None, height=None, units='in', dpi=None, limitsize=True, **kwargs)
- Save a ggplot object as an image file

Parameters:
- filename (str or file) – File name or file to write the plot to.
- format (str) – Image format to use, automatically extract from file name extension.
- path (str) – Path to save plot to (if you just want to set path and not filename).
- width (number) – Width (defaults to the width of current plotting window).
- height (number) – Height (defaults to the height of current plotting window).
- units (str) – Units for width and height when either one is explicitly specified (in, cm, or mm).
- dpi (float) – DPI to use for raster graphics.   
  If None, defaults to using the dpi of theme, if none is set then a dpi of 100.
- limitsize (bool) – If True (the default), ggsave will not save images larger than 50x50 inches,   
  to prevent the common error of specifying dimensions in pixels.  
- kwargs (dict) – Additional arguments to pass to matplotlib savefig().
